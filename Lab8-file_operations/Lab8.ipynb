### 04) WAP to find the longest word in a file named abc.txt

```python

f = open("abc.txt")

content = f.read()
f.close()

words = content.split()

longest=''
max=0

for i in words:
    if len(i) > max:
        longest = i
        max = len(i)
print(longest)
print("length:",max)
```

### 01) WAP to write first 100 prime numbers to a file named primenumbers.txt 
(Note: each number should be in new line)

```python
def isPrime(i):
    for j in range(2,i):
        if i % j == 0:
            return False
    return True
# print(isPrime(9))

count = 0
i = 2

f = open("primenumbers.txt","w")
while count < 100:
    if isPrime(i):
        f.write(str(i)+'\n')
        count+=1
    i+=1
f.write("Total:"+str(count))
```

### 02) WAP to merge two files and write it in a new file.

```python
file1 = "file1.txt"
file2 = "file2.txt"
merged_file = "merged_file.txt"

f1 = open(file1,"r")
f2 = open(file2,"r")  

f3 = open(merged_file,"w")    

for line in f1:
    f3.write(line)

for line in f2:
    f3.write(" + "+line+" = combineFileData")
   
f1.close()  
f2.close()
f3.close()

print("Merged file1 and file2 into",merged_file)
```

### 03) WAP to encrypt a text file.

```python
file=open("abc.txt","r+")
data=file.read()
print(data)

key=5
file2=open("encrypt.txt","w")

for i in data:
    print(i)
    b=ord(i)+key
    file2.write(chr(b))
file2.close()
    
fileopen=open("encrypt.txt","r")
data=fileopen.read()
print(data)
```

### 04) WAP to decrypt a previously encrypted file.

```python
file=open("encrypt.txt","r")
data=file.read()
# print(data)

key=5

file2=open("decrypt.txt","w")

for i in data:
    b=ord(i)-key
    file2.write(chr(b))
file2.close()


file=open("decrypt.txt","r")
data1=file.read()
print(data1)
```
