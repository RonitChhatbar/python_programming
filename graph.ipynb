```python
x = [25,25,25]

l = ['p1','p2','p3']
c = ['r','c','y']
e = [0,0.1,0]

plt.pie(x,colors=c,labels=l,explode=e,shadow=True,autopct='%1.1f%%')
```

```python
r.seed(1)

x = [r.randrange(1,11) for _ in range(1,4)]

l = ['1','2','3']
w = [0.4,0.2,0.5]
c = ['r','m','c']

plt.title("data")
plt.bar(x,width=w,label=l,height=1,color=c)
```

```python
import networkx as n
g = n.Graph()
g.add_edge('path1','path2')
g.add_edge('path2','path3')
g.add_edge('path3','path1')

n.draw(g,with_labels=True)
```

```python
import networkx as n
g = n.DiGraph()#DIRECTED GRAPH
g.add_edge('path1','path2')
g.add_edge('path2','path3')
g.add_edge('path3','path1')

n.draw(g,with_labels=True)
```
