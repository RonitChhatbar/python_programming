### 01) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
product = ['a','b','c','d']
price = [33,24,51,15]

plt.bar(product,price,color='black')
for product,price in zip(product,price):
    plt.text(product,price,str(price),ha="center",va="bottom")
plt.xlabel('product')
plt.ylabel('price')

```

### 02) WAP create a Scatter Plot with several colors in Matplotlib?

```python
x = [r.randrange(1,20) for _ in range(1,51)]
y = [r.randrange(1,20) for _ in range(1,51)]
color = [r.randrange(1,20) for _ in range(1,51)]
plt.scatter(x,y,c=color)
plt.colorbar()
```
